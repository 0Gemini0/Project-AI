```python
data['action'] = data['Stimulus'].apply(lambda x: x.replace(' ', '.').split('.')[0])
data2['action'] = data2['Stimulus'].apply(lambda x: x.replace(' ', '.').split('.')[0])
data3['action'] = data3['Stimulus'].apply(lambda x: x.replace(' ', '.').split('.')[0])
data3.head()
```

```python
def plot_confidence(data):
    avg = np.mean(data, axis = 0)
    sd = np.std(data, axis = 0)
    
    plt.figure()
    plt.plot(avg)
    plt.fill_between(list(range(len(avg))), avg - sd, avg + sd, alpha=.3, facecolor='red')
    
def pre_analysis(data, action, repetitions, wave, stimulus_len = None):
    dt = data[wave]
    dt = dt[data.action == action].copy()
    
    if stimulus_len is None:
        stimulus_len = int(dt.shape[0] / repetitions)
        
    datamatrix = np.zeros([repetitions, stimulus_len])
    for i in range(repetitions):
        datamatrix[i, :] = dt[i*stimulus_len:(i+1)*stimulus_len]
    
    plot_confidence(datamatrix)
    
```

```python
pre_analysis(data, "run", 10, "AF3")

pre_analysis(data, "walk", 10, "AF3")

pre_analysis(data, "jump", 10, "AF3")

pre_analysis(data, "crouch", 10, "AF3")
```

```python
data.fillna(0)
data2.fillna(0)
data3.fillna(0)
```

```python
run = np.asarray(data[data.action=='run'].iloc[:,1:-1])
walk = np.asarray(data[data.action=='walk'].iloc[:,1:-1])
crouch = np.asarray(data[data.action=='crouch'].iloc[:,1:-1])
jump = np.asarray(data[data.action=='jump'].iloc[:,1:-1])
run2 = np.asarray(data2[data2.action=='run'].iloc[:,2:-1])
walk2 = np.asarray(data2[data2.action=='walk'].iloc[:,2:-1])
crouch2 = np.asarray(data2[data2.action=='crouch'].iloc[:,2:-1])
jump2 = np.asarray(data2[data2.action=='jump'].iloc[:,2:-1])
run3 = np.asarray(data3[data3.action=='run'].iloc[:,2:-1])
walk3 = np.asarray(data3[data3.action=='walk'].iloc[:,2:-1])
crouch3 = np.asarray(data3[data3.action=='crouch'].iloc[:,2:-1])
jump3 = np.asarray(data3[data3.action=='jump'].iloc[:,2:-1])
run.shape, walk.shape, crouch.shape, jump.shape,run2.shape, walk2.shape, crouch2.shape, jump2.shape,run3.shape, walk3.shape, crouch3.shape, jump3.shape
```

```python
len_signal = 280
X_train = np.zeros((80,len_signal,14))
y_train = np.zeros((80))
X_test = np.zeros((40,len_signal,14))
y_test = np.zeros((40))
```

```python
#first set
n_start = 0
n_end = 280
for i in range(40):
    print(i)
    if i%4 == 0: #run 
        for j in range(14):
            X_train[i,0:len_signal,j] = run[n_start:n_end,j]
            y_train[i] = 0
    if i%4 == 1: #walk
         for j in range(14):
            X_train[i,0:len_signal,j] = walk[n_start:n_end,j]
            y_train[i] = 1
    if i%4 == 2: #crouch
         for j in range(14):
            X_train[i,0:len_signal,j] = crouch[n_start:n_end,j]
            y_train[i] = 2
    if i%4 == 3: #jump
        for j in range(14):
            X_train[i,0:len_signal,j] = jump[n_start:n_end,j]
            y_train[i] = 3    
        n_start += 280
        n_end += 280
```

```python
X_train.shape
```

```python
n_start = 0
n_end = 280
for i in range(40,80):
    print(i)
    if i%4 == 0: #run 
        for j in range(14):
            X_train[i,0:len_signal,j] = run2[n_start:n_end,j]
            y_train[i] = 0
    if i%4 == 1: #walk
         for j in range(14):
            X_train[i,0:len_signal,j] = walk2[n_start:n_end,j]
            y_train[i] = 1
    if i%4 == 2: #crouch
         for j in range(14):
            X_train[i,0:len_signal,j] = crouch2[n_start:n_end,j]
            y_train[i] = 2
    if i%4 == 3: #jump
        for j in range(14):
            X_train[i,0:len_signal,j] = jump2[n_start:n_end,j]
            y_train[i] = 3    
        n_start += 280
        n_end += 280    
```

```python
X_train.shape
```

```python
#test set
n_start = 0
n_end = 280
for i in range(40):
    print(i)
    if i%4 == 0: #run 
        for j in range(14):
            X_test[i,0:len_signal,j] = run3[n_start:n_end,j]
            y_test[i] = 0
    if i%4 == 1: #walk
         for j in range(14):
            X_test[i,0:len_signal,j] = walk3[n_start:n_end,j]
            y_test[i] = 1
    if i%4 == 2: #crouch
         for j in range(14):
            X_test[i,0:len_signal,j] = crouch3[n_start:n_end,j]
            y_test[i] = 2
    if i%4 == 3: #jump
        for j in range(14):
            X_test[i,0:len_signal,j] = jump3[n_start:n_end,j]
            y_test[i] = 3    
        n_start += 280
        n_end += 280
```

```python
y_train.shape, X_train.shape, y_test.shape, X_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.preprocessing import sequence
# fix random seed for reproducibility
np.random.seed(7)

# truncate and pad input sequences
X_train = sequence.pad_sequences(X_train, maxlen=len_signal)
X_test = sequence.pad_sequences(X_test,maxlen=len_signal)
```

```python
model = Sequential()
model.add(LSTM(64, input_shape=(280,14)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=1, verbose = 2)
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
