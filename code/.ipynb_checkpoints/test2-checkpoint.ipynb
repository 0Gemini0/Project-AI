```python
import numpy as np
import scipy as sp
from scipy import fftpack

import pandas as pd

import os
import glob

import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
path = os.getcwd() + '\\data\\'
extension = 'csv'

os.chdir(path)
titles = glob.glob('*.{}'.format(extension))
```

```python
datasets = []

for d in titles:
    datasets.append(pd.read_csv(path + d))
```

```python
len(datasets)
```

```python
data = datasets[0]
print(data)
data = data[data.columns[1:]]
list(data.columns)
```

```python
data['action'] = data['Stimulus'].apply(lambda x: x.replace(' ', '.').split('.')[0])
data.head()
```

```python
sb.set_context("poster")
for c in data.columns[1:-1]:
    plt.figure()
    sb.violinplot(data[c], data.action)
```

```python
def plot_confidence(data):
    avg = np.mean(data, axis = 0)
    sd = np.std(data, axis = 0)
    
    plt.figure()
    plt.plot(avg)
    plt.fill_between(list(range(len(avg))), avg - sd, avg + sd, alpha=.3, facecolor='red')
    
def pre_analysis(data, action, repetitions, wave, stimulus_len = None):
    dt = data[wave]
    dt = dt[data.action == action].copy()
    
    if stimulus_len is None:
        stimulus_len = int(dt.shape[0] / repetitions)
        
    datamatrix = np.zeros([repetitions, stimulus_len])
    for i in range(repetitions):
        datamatrix[i, :] = dt[i*stimulus_len:(i+1)*stimulus_len]
    
    plot_confidence(datamatrix)
    
```

```python
pre_analysis(data, "run", 10, "AF3")

pre_analysis(data, "walk", 10, "AF3")

pre_analysis(data, "jump", 10, "AF3")

pre_analysis(data, "crouch", 10, "AF3")
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier

inp = np.array([[0.1,0.2],[0.9,0.8], [0.3, 0.4], [0.8, 0.7]])
out = np.array([0,1,0,1])

def classify_net(inp, out):
    clf = MLPClassifier(solver="lbfgs", alpha=1e-4,hidden_layer_sizes=(200, 100, 50, 50),random_state=1)
    print(clf.fit(inp, out))
    print(clf.predict(inp))
    print(clf.predict_proba(inp))
    
def classify_knn(inp, out):
    clf = KNeighborsClassifier(n_neighbors=3)
    print(clf.fit(inp, out))
    print(clf.predict(inp))
    print(clf.predict_proba(inp))
    
    
classify_net(inp,out)
classify_knn(inp,out)
    
```
